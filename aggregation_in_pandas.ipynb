**Aggregation **- computing a summary statistic about each group,
*   compute group sums or means
*   compute group sizes/counts

Common aggregation functions: min,max,count,sum, prod,mean,median, mode,mad,std,var



```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Salaries.csv')
```

```python
#agg() method are useful when multiple statistics are computed per column:
df[['phd','salary']].agg(['min','mean','max'])
```

# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#univariate - "one variable " dta visualization

```

```python
#strip Plot
sns.stripplot(y = "tip", data =tips, jitter = True)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(y='mpg',data=mtcars,jitter=True)

plt.show()
```

```python
#grouping with stripplot
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('Tips($)')
plt.xlabel('Days')
plt.show()
```

```python
sns.stripplot(x='gear',y='cyl',data=mtcars)
```

```python
sns.stripplot(x='day',y='tip',data=tips, jitter=False)
plt.ylabel('Tips($)')
plt.xlabel('Days')
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tips($)')
plt.show()
```

```python
#changing orientation
sns.swarmplot(x='tip',y='day',data=tips,orient='h',hue='sex')
plt.xlabel('tips($)')
plt.show()
```

# Box and Violin plot

```python
#plt.subplot(1,2,3)
sns.boxplot(x='day',y='tip',data= tips)
plt.ylabel('tip($)')
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
#Combining Plots

sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgray')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

# Bivariate - "two variables",data visualization

```python
#joint Plot
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#Density Plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
#pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
