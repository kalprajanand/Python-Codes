Data frames *Groupby* Method **bold text**

using "groupby" method we can:
 
1.  split the data into groups based on some criteria 
2.  calculate statistics(or apply a functions) to each group
3. similar dplyr()function in R

```python
# Group dat using rank{groupby function is to group data having same values}
df_rank = df.groupby(['rank'])
```

```python
df_rank.mean()
```

```python
#calculate mean salary for each professor rank:
df.groupby('rank')[('salary')].mean()
#Note: if single brackets are used to specify the column(e.g salary), then the 
#output is Pandas Series object. 
#when double brackets are used the output is a data frame.
```
